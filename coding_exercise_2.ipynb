```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://www.nfl.com/standings/league/2019/REG'
page = requests.get(url)
if page.status_code == 200:
    print("Can scrape")
else:
    print('Not allowed')
soup = BeautifulSoup(page.text,'lxml')
```

```python
import re
```

```python
table = soup.find('table',class_ = re.compile('d3-o-table d3-o-table--row-'))
```

```python
headers = [i.text.strip() for i in table.find_all('th')]
```

```python
stat_table = pd.DataFrame(columns=headers)
```

```python
rows = table.find_all('tr')[1:][::-1]
```

```python
for k in rows:
    stat_table = stat_table.append(dict(zip(headers,[i.text.strip() if j != 0 else i.find('div',class_='d3-o-club-fullname').text.strip() for j,i in enumerate(k.find_all('td'))])),ignore_index=True)
```

```python
for k in rows:
    stat_table = stat_table.append(dict(zip(headers,[i.text.strip() if j != 0 else " ".join(i.find('div',class_='d3-o-club-fullname').text.split(" ")[:-1])  for j,i in enumerate(k.find_all('td'))])),ignore_index=True)
```
